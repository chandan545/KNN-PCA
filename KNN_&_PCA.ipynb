Theoretical

1. What is K-Nearest Neighbors (KNN) and how does it work?
   - K-Nearest Neighbors (KNN) is a simple machine learning algorithm used for classification and regression.
   - How it works:
     - It stores all training data.
     - When a new data point comes in, it finds the K closest data points (neighbors) from the training set.
     - For classification, it assigns the most common class among the K neighbors.
     - For regression, it averages the values of the K neighbors.
   - Itâ€™s based on the idea that similar things exist near each other.

2. What is the difference between KNN Classification and KNN Regression?
   - KNN Classification
     - Predicts: Class/Category (e.g., "Yes" or "No")
     - Method: Majority vote among K nearest neighbors
     - Used for: Tasks like spam detection, disease prediction
     - Output type: Discrete (labels)

   - KNN Regression
     - Predicts: Numerical value (e.g., 45.6, 102.0)
     - Method: Average of values from K nearest neighbors
     - Used for: Tasks like price prediction, temperature forecasting
     - Output type: Continuous (numbers)
   
3. What is the role of the distance metric in KNN?
   - The distance metric determines how close or similar data points are to each other. It helps the algorithm find the K nearest neighbors to the new data point.
   - Important Points:
     - Common distance metrics: Euclidean, Manhattan, Minkowski
     - A good metric improves the accuracy of predictions
     - It directly affects the neighbors selected and the final output
   - In short: The distance metric is key to measuring similarity in KNN.

4. What is the Curse of Dimensionality in KNN?
   - It refers to the problem where KNN performance drops as the number of features (dimensions) increases.
   - Important Points:
     - In high dimensions, data points become sparse and less similar to each other
     - Distance metrics become less meaningful, making it hard to find true neighbors
     - It can lead to low accuracy and longer computation time
   - In short: More dimensions = harder for KNN to find reliable neighbors.

5. How can we choose the best value of K in KNN?
   - We choose the best K by trying different values and checking which gives the highest accuracy.
   - Important Points:
     - Use cross-validation to test multiple K values
     - A small K (e.g., 1) may lead to overfitting
     - A large K may lead to underfitting
     - Common practice: try odd numbers (to avoid ties)
   - In short: Try different K values using validation to find the one that gives the best performance.

6. What are KD Tree and Ball Tree in KNN?
   - KD Tree and Ball Tree are data structures used to speed up the nearest neighbor search in KNN.
   - KD Tree (K-Dimensional Tree):
     - A binary tree structure that splits data points along each axis in the feature space (one at a time).
     - Best suited for low to medium dimensional spaces.
     - Faster than brute-force search for low dimensions, as it reduces the number of points to check.
   - Ball Tree:
     - Organizes data points in "balls" (regions in feature space) instead of using axis-aligned splits.
     - Better for higher-dimensional spaces compared to KD Tree.
     - Works by partitioning data using a recursive ball of points, improving efficiency in high dimensions.
   - Key Differences:
     - KD Tree: Best for low dimensions, splits data along axes.
     - Ball Tree: Best for high dimensions, uses ball-like regions.
   - In short, both trees optimize the KNN search but work better in different conditions.

7. When should you use KD Tree vs. Ball Tree?
   - Use KD Tree when:
     - The dataset has low to medium dimensions (typically less than 20-30).
     - You need to speed up nearest neighbor searches in 2D, 3D, or other low-dimensional spaces.
     - The data is axis-aligned and well-suited to partitioning along axes.
   - Use Ball Tree when:
     - The dataset has high dimensions (greater than 30).
     - The data is non-axis-aligned or has complex relationships, making it harder to split using axis-aligned planes.
     - You need to efficiently handle sparse data or data with irregular distributions.
   - In short:
     - KD Tree is better for low-dimensional spaces.
     - Ball Tree is better for high-dimensional spaces.

8. What are the disadvantages of KNN?
   - Disadvantages of KNN:
     - Computationally Expensive:
       - KNN requires calculating distances between the query point and all training points, making it slow for large datasets, especially during prediction.
     - Sensitive to Irrelevant Features:
       - KNN performs poorly when the dataset contains irrelevant or noisy features, as the distance metric may be affected.
     - High Memory Usage:
       - KNN stores the entire training dataset in memory, which can be problematic for large datasets.
     - Curse of Dimensionality:
       - As the number of features increases, the distance between points becomes less meaningful, leading to poor performance in high-dimensional spaces.
     - Choice of K:
       - The performance of KNN heavily depends on choosing the correct value of K. An incorrect K can lead to overfitting (small K) or underfitting (large K).
     - Not Suitable for Imbalanced Datasets:
       - KNN may struggle with imbalanced datasets where certain classes are underrepresented, as the nearest neighbors may belong predominantly to the majority class.
   - In short: KNN can be slow, memory-intensive, and sensitive to irrelevant features, especially with large or high-dimensional data.

9. How does feature scaling affect KNN?
   - Feature Scaling and Its Effect on KNN:
     - Feature scaling is crucial in KNN because it directly impacts how distances between data points are calculated.
   - How it Affects KNN:
     - Unequal Contribution of Features:
       - Without scaling, features with larger ranges (e.g., age in years vs. income in thousands) dominate the distance calculation. This can lead to misleading results, as the model may prioritize certain features over others.
     - Distance Calculation:
       - KNN relies on distance metrics like Euclidean distance, which is sensitive to the scale of features. Without scaling, large-scale features can skew the distance and affect the selection of nearest neighbors.
     - Improved Accuracy:
       - Scaling (e.g., using Standardization or Normalization) ensures all features contribute equally to distance calculations, leading to more accurate results.
     - Common Scaling Techniques:
       - Standardization (Z-score normalization): Centers the data around 0 with a standard deviation of 1.
       - Min-Max Scaling: Rescales features to a range between 0 and 1.

10. What is PCA (Principal Component Analysis)?
    - Principal Component Analysis (PCA) is a dimensionality reduction technique used to simplify large datasets while retaining as much information as possible.
    - Key Points:
      - Reduces Dimensions: Transforms data into fewer dimensions (principal components) that retain the most variance.
      - Uncorrelated Features: The principal components are orthogonal (uncorrelated) to each other.
      - Variance Maximization: The first principal component captures the most variance, the second captures the next most, and so on.
      - Improves Performance: Reduces noise and improves computation time in machine learning models.
      - Steps:
        - Standardize the data.
        - Compute the covariance matrix.
        - Extract eigenvalues and eigenvectors.
        - Project data onto principal components.
      - In short: PCA reduces data dimensions, enhances model performance, and helps in visualizing complex data by focusing on the most important patterns.

11. How does PCA work?
    - How PCA Works:
      - Standardize the Data:
        - If the data features have different scales, first standardize them (mean = 0, standard deviation = 1) so they contribute equally.
      - Compute the Covariance Matrix:
        - This matrix represents how features in the dataset are correlated with each other. It captures the relationships between different features.
      - Calculate Eigenvalues and Eigenvectors:
        - The eigenvectors represent the directions of maximum variance in the data, and the eigenvalues indicate how much variance is captured by each eigenvector.
      - Sort Eigenvectors:
        - Sort the eigenvectors based on their eigenvalues (in descending order). The larger the eigenvalue, the more variance the corresponding eigenvector explains.
      - Select Principal Components:
        - Choose the top K eigenvectors (principal components) based on the sorted eigenvalues. These components will form a new feature space with reduced dimensions.
      - Project Data onto New Feature Space:
        - Multiply the original data matrix by the selected eigenvectors to transform the data into a lower-dimensional space.
      - In short: PCA works by transforming the data into new axes (principal components) that capture the maximum variance, reducing the datasetâ€™s dimensionality while preserving key patterns.

12. What is the geometric intuition behind PCA?
    - Geometric Intuition Behind PCA:
      - PCA can be understood as a way to find the "best-fit" line or plane (or hyperplane in higher dimensions) that represents the data in a lower-dimensional space.
    - Key Geometric Concepts:
      - Maximizing Variance:
        - PCA finds new axes (principal components) that capture the maximum variance of the data. The first principal component (PC1) is the direction that has the largest spread (variance) of the data points. The second principal component (PC2) is orthogonal to PC1 and captures the second-largest variance, and so on.
     - Projection onto New Axes:
       - The original data points are projected onto these new axes (principal components). This projection helps reduce the dimensions of the data while retaining as much information (variance) as possible.
     - Data Centering:
       - The data is centered around the mean of the dataset (subtracting the mean of each feature), so the new axes start from the "center" of the data cloud. This ensures the directions of maximum variance are well-defined.
     - Orthogonality of Components:
       - The principal components are orthogonal (perpendicular) to each other, ensuring that each component captures a unique aspect of the data's variance without overlap.
     - Visualizing in 2D (Simple Example):
       - Imagine a cloud of data points spread across a 2D plane. PCA identifies the line (1st principal component) that best represents the direction of maximum spread in the data.
       - The 2nd principal component is a line perpendicular to the first, capturing the variance in the remaining direction.
       - By projecting the data onto these two lines (or axes), we reduce the data from 2D to 1D (if only the first principal component is kept), retaining the most important information.
     - In short: PCA geometrically finds the directions of maximum variance in the data and projects the data onto these directions, reducing dimensions while preserving key patterns.

13. What is the difference between Feature Selection and Feature Extraction?
    - Feature Selection:
      - Method: Selects a subset of the original features without changing them.
      - Goal: Retains only the most important features.
      - Example: Removing irrelevant or redundant features.
    - Feature Extraction:
      - Method: Transforms the original features into a new set of features (usually lower-dimensional).
      - Goal: Create new features that summarize the original data.
      - Example: Using PCA to create new principal components.
    - In short: Feature selection picks important features, while feature extraction creates new, reduced features.

14. What are Eigenvalues and Eigenvectors in PCA?
    - In PCA, eigenvalues and eigenvectors are fundamental concepts that help transform the data into a new set of uncorrelated features (principal components).
    - Eigenvectors:
      - Definition: Eigenvectors represent the directions in the feature space along which the data has the most variance.
      - Role in PCA: They are the new axes (principal components) onto which the original data is projected. Each eigenvector corresponds to one principal component.
      - Geometrically: The eigenvector is the line or plane that best fits the data in a given direction of maximum variance.
   - Eigenvalues:
      - Definition: Eigenvalues measure the magnitude or amount of variance explained by the corresponding eigenvector.
      - Role in PCA: The eigenvalue tells us how much variance (information) is captured by the eigenvector. Larger eigenvalues mean that the corresponding eigenvector (principal component) explains more variance in the data.
      - Geometrically: The eigenvalue indicates how spread out the data is along that principal component (axis).
   - Summary:
      - Eigenvectors give us the directions of maximum variance.
      - Eigenvalues tell us how much variance each direction explains.
   - In short, eigenvectors define the new axes, and eigenvalues tell us how important those axes are in capturing the data's variability.

15. How do you decide the number of components to keep in PCA?
    - To decide the number of components to keep in PCA, you can use the following methods:
    - Explained Variance Ratio:
      - Goal: Keep enough components to capture a high percentage of the total variance in the data.
      - How: Look at the cumulative explained variance (sum of eigenvalues) and choose the smallest number of components that explain a desired percentage of the total variance, often 90-95%.
      - Example: If the first 3 components explain 92% of the variance, you might choose to keep those 3 components.
    - Scree Plot:
      - Goal: Identify the "elbow" point where the explained variance starts to level off.
      - How: Plot the eigenvalues or the explained variance of each principal component. The "elbow" point on the plot indicates where adding more components provides diminishing returns in variance explained.
      - Example: The point where the curve flattens is often a good place to cut off additional components.
    - Kaiser Criterion (Eigenvalue > 1):
      - Goal: Retain components with eigenvalues greater than 1.
      - How: Only keep components that have an eigenvalue greater than 1, as they explain more variance than a single feature would.
    - In short: Choose the number of components that explain a significant portion of the variance (e.g., 90-95%) or look for the "elbow" in a scree plot to determine when additional components no longer add much value.

16. Can PCA be used for classification?
    - Yes, PCA can be used for classification, but not directly as a classification algorithm. Instead, it is typically used as a preprocessing step to reduce the dimensionality of the data before applying a classification algorithm.
    - How PCA Helps in Classification:
      - Dimensionality Reduction:
        - PCA reduces the number of features (dimensions) in the dataset while retaining the most important information, making the dataset easier to handle and speeding up the training of classification models.
      - Noise Reduction:
        - By keeping only the most important principal components, PCA can help remove noisy or irrelevant features, improving the performance of classifiers.
      - Visualization:
        - PCA can also be used to visualize high-dimensional data in 2D or 3D, helping in understanding patterns and relationships that may aid in classification tasks.
      - Workflow:
        - Apply PCA to reduce the dimensionality of the data.
        - Use the reduced set of principal components as input to a classification algorithm (e.g., Logistic Regression, SVM, KNN).

17. What are the limitations of PCA?
    - Limitations of PCA:
      - Linear Assumption:
        - PCA assumes that the data lies on a linear subspace. It may not perform well on datasets with complex, non-linear relationships.
      - Sensitivity to Scaling:
        - PCA is sensitive to the scaling of features. Features with larger ranges dominate the principal components unless data is standardized.
      - Interpretability:
        - The new principal components are linear combinations of the original features, making them harder to interpret compared to the original features.
      - Loss of Information:
        - When reducing dimensions, some information may be lost, especially if too few components are kept, which could degrade model performance.
      - Assumption of Gaussian Distribution:
        - PCA works best when the data is normally distributed. It may not capture important structures in data that do not follow this assumption.
      - Outlier Sensitivity:
        - PCA is sensitive to outliers, as they can disproportionately influence the principal components, leading to skewed results.
      - Computational Complexity:
        - For very large datasets, computing PCA can be computationally expensive because it requires calculating eigenvalues and eigenvectors, which can be slow for high-dimensional data.

18. How do KNN and PCA complement each other?
    - KNN (K-Nearest Neighbors) and PCA (Principal Component Analysis) can complement each other in machine learning by combining dimensionality reduction with distance-based classification.
    - How they complement each other:
    - Dimensionality Reduction with PCA:
      - PCA reduces the number of features in the dataset while retaining most of the variance (information). This helps simplify the data and removes noise, improving the efficiency of the KNN algorithm.
      - By reducing dimensions, PCA makes the distance calculations in KNN more meaningful, especially when dealing with high-dimensional data.
    - Improved KNN Performance:
      - KNN can suffer from the curse of dimensionality in high-dimensional spaces, where distances between points become less informative. PCA addresses this by transforming the data into a lower-dimensional space, making it easier for KNN to find meaningful neighbors.
      - With fewer dimensions, the computational cost for finding the nearest neighbors is reduced, speeding up KNN.
    - Noise Reduction:
      - PCA helps remove less important features (those with low variance) that may be noisy or irrelevant. This results in a more accurate KNN model, as KNN will focus on the most important features for classification.
    - Workflow:
      - Apply PCA to reduce the dimensions of the dataset.
      - Use the transformed data (reduced features) as input to the KNN classifier.
    - In short: PCA helps by reducing dimensions and removing noise, making the distance calculations in KNN more effective and efficient, leading to improved performance in classification tasks.

19. How does KNN handle missing values in a dataset?
    - KNN doesn't handle missing values directly, but you can use these strategies before applying KNN:
    - Imputation:
      - Replace missing values with mean/median (for numerical data) or mode (for categorical data).
      - Use KNN imputation to predict missing values based on nearest neighbors.
    - Remove Data Points:
      - Remove rows with missing values, but this may lead to data loss.
    - Distance-based Imputation:
      - Compute distances ignoring missing values or treat them as missing for distance calculation.
    - Use Algorithms that Handle Missing Data:
      - Some KNN variants can handle missing values through partial distance metrics or weighted neighbors.

20. What are the key differences between PCA and Linear Discriminant Analysis (LDA)?
   - Here are the key differences between PCA (Principal Component Analysis) and LDA (Linear Discriminant Analysis):
   - **Purpose**:
     - PCA:
          - Unsupervised technique used for dimensionality reduction.
          - Aims to maximize variance in the data and project the data into lower dimensions.
     - LDA:
          - Supervised technique used for both dimensionality reduction and classification.
          - Aims to maximize the separation between different classes while minimizing variance within each class.
   - **Objective**:
    - PCA:
         - Seeks to find the directions (principal components) that capture the most variance in the data.
    - LDA:
         - Seeks to find a linear combination of features that best separates the classes in the dataset.
   - **Data Type**:
    - PCA:
         - Can be used on unlabeled data because it doesnâ€™t consider class labels.
    - LDA:
         - Requires labeled data because it uses class labels to maximize class separability.
   - **Focus**:
    - PCA:
        - Focuses on preserving the overall variance in the data.
    - LDA:
        - Focuses on maximizing class separability by emphasizing the differences between classes.
   - **Output**:
    - PCA:
       - Projects data into orthogonal components based on variance.
    - LDA:
       - Projects data into a lower-dimensional space while maximizing class separation.
   - **Number of Components**:
    - PCA:
       - Can reduce to as many dimensions as the original data (or fewer, depending on variance retention).
    - LDA:
       - The maximum number of components is C-1 (where C is the number of classes) since it tries to maximize class separability.
    - **Summary**:
       - PCA is an unsupervised technique for reducing dimensions by preserving variance.
       - LDA is a supervised technique for reducing dimensions by maximizing class separability.

```python
#1. Train a KNN Classifier on the Iris dataset and print model accuracy.
#ans.
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_train, y_train)

# Predict on the test set
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy}")
```

```python
#4. Train a KNN Classifier with different values of K and visualize decision boundaries
#ans.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for easier visualization
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a function to plot decision boundaries
def plot_decision_boundary(clf, X, y, ax):
    h = 0.02  # step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    ax.set_title(f"KNN (K={clf.n_neighbors})")
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")

# Create subplots to visualize decision boundaries for different K values
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Train and plot decision boundaries for different values of K
for i, k in enumerate([1, 5, 15]):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    plot_decision_boundary(knn, X, y, axs[i])

plt.tight_layout()
plt.show()
```

```python
#5. Apply Feature Scaling before training a KNN model and compare results with unscaled data.
#ans.
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for easier visualization
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Apply feature scaling using StandardScaler
scaler = StandardScaler()

# Scale the training and test sets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train KNN model on unscaled data
knn_unscaled = KNeighborsClassifier(n_neighbors=5)
knn_unscaled.fit(X_train, y_train)
y_pred_unscaled = knn_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

# Train KNN model on scaled data
knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

# Print the results
print(f"Accuracy on unscaled data: {accuracy_unscaled}")
print(f"Accuracy on scaled data: {accuracy_scaled}")
```

```python
#8. Perform Hyperparameter Tuning on a KNN Classifier using GridSearchCV.
#ans.
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize KNN Classifier
knn = KNeighborsClassifier()

# Define hyperparameter grid for tuning
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Initialize GridSearchCV with KNN and parameter grid
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit GridSearchCV
grid_search.fit(X_train, y_train)

# Best parameters and best score
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)

# Evaluate the best model on the test set
best_knn = grid_search.best_estimator_
test_accuracy = best_knn.score(X_test, y_test)
print("Test Accuracy with Best Model: ", test_accuracy)
```

```python
#18. Train a PCA model and analyze the effect of different numbers of components on accuracy.
#ans.
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data (important for PCA)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# List to store accuracy values for different number of components
accuracy_values = []

# Loop over a range of principal components
for n_components in range(1, X_train.shape[1] + 1):
    # Apply PCA
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    # Train KNN Classifier on the reduced data
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train_pca, y_train)

    # Predict and calculate accuracy
    y_pred = knn.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)

    # Store the accuracy for this number of components
    accuracy_values.append(accuracy)

# Plot the effect of the number of components on accuracy
plt.figure(figsize=(8, 6))
plt.plot(range(1, X_train.shape[1] + 1), accuracy_values, marker='o', linestyle='-', color='b')
plt.title('Effect of Number of PCA Components on Accuracy')
plt.xlabel('Number of PCA Components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# Print the accuracy values
for n_components, accuracy in zip(range(1, X_train.shape[1] + 1), accuracy_values):
    print(f"Number of Components: {n_components}, Accuracy: {accuracy}")
```

```python
#23. Train a KNN Classifier and evaluate using ROC-AUC score.
#ans.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import label_binarize

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Predict the probabilities of the classes
y_prob = knn.predict_proba(X_test_scaled)

# Binarize the output labels for multi-class ROC-AUC computation (One-vs-Rest strategy)
y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

# Compute the ROC-AUC score for each class
roc_auc = roc_auc_score(y_test_bin, y_prob, multi_class='ovr')

# Print the ROC-AUC score
print(f"ROC-AUC score: {roc_auc}")

# Optional: Plot the ROC curve for each class
from sklearn.metrics import roc_curve

# Plot ROC curve for each class
plt.figure(figsize=(10, 6))

for i in range(3):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc_score(y_test_bin[:, i], y_prob[:, i]):.2f})')

plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.title('ROC Curve for KNN Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
#25. Train a KNN Classifier and perform feature selection before training.
#ans.
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Perform feature selection (Select top 2 features)
selector = SelectKBest(score_func=f_classif, k=2)
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Train the KNN Classifier on the selected features
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_selected, y_train)

# Predict on the test set
y_pred = knn.predict(X_test_selected)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
#26. Train a PCA model and visualize the data reconstruction error after reducing dimensions.
#ans.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the data (important for PCA)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train the PCA model
pca = PCA(n_components=2)  # Reduce the data to 2 components for visualization
X_pca = pca.fit_transform(X_scaled)

# Reconstruct the data from the reduced components
X_reconstructed = pca.inverse_transform(X_pca)

# Calculate the reconstruction error (MSE)
reconstruction_error = mean_squared_error(X_scaled, X_reconstructed)
print(f"Reconstruction Error (MSE): {reconstruction_error}")

# Visualize the data before and after PCA (scatter plot)
plt.figure(figsize=(10, 6))

# Scatter plot of original data
plt.subplot(1, 2, 1)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='blue', alpha=0.7, label="Original Data")
plt.title("Original Data (First 2 Features)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

# Scatter plot of reconstructed data
plt.subplot(1, 2, 2)
plt.scatter(X_reconstructed[:, 0], X_reconstructed[:, 1], color='red', alpha=0.7, label="Reconstructed Data")
plt.title("Reconstructed Data After PCA")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#27. Train a KNN Classifier and visualize the decision boundary.
#ans.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Use only the first two features for 2D visualization
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the data (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

# Create a meshgrid for plotting decision boundary
h = .02  # Step size in the mesh
x_min, x_max = X_train_scaled[:, 0].min() - 1, X_train_scaled[:, 0].max() + 1
y_min, y_max = X_train_scaled[:, 1].min() - 1, X_train_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Predict class for each point in the meshgrid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=y_train, cmap=plt.cm.coolwarm, edgecolors='k', marker='o', s=50, label='Train Data')
plt.scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], c=y_test, cmap=plt.cm.coolwarm, edgecolors='k', marker='x', s=50, label='Test Data')
plt.title('KNN Classifier - Decision Boundary')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.legend()
plt.grid(True)
plt.show()
```
